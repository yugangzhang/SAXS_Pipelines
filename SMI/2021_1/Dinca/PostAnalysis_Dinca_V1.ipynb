```python
sams = ['_Dinca_Zn_PDI.h5', '_Dinca_Mn_PDI.h5', '_Dinca_PDI_Ligand.h5',
       '_Dinca_RW_1315.h5', '_Dinca_Zn_PDI_2.h5', '_Dinca_Mn_PDI_32.h5',
       '_Dinca_NiBHT.h5', '_Dinca_NiBHT_Neg.h5', '_Dinca_NiBHT_Pos.h5',
       '_Dinca_TIBQ_COF_DMA.h5', '_Dinca_TIBA_COF_CTAB.h5']
sam = sams[0]
```

```python
def _get_saxs():
    qiq_dict = {}
    for sam in sams:
        Res = h5todict( inDir + sam  )['SAXS']
        fps = Res['md']['filenames']
        fp = fps[0]
        print( fp )
        qiq_dict[ sam ] =Res[fp]['cir_avg']['data']
    return qiq_dict    
        
def _get_waxs():
    qiq_dict = {}
    for sam in sams:
        Res = h5todict( inDir + sam  )['WAXS']
        #fps = Res['md']['filenames']
        #fp = fps[0]
        #print( fp )
        qiq_dict[ sam ] =Res['cir_avg']['data']   
    return qiq_dict        
```

```python
qiqs = _get_saxs()
qiqw = _get_waxs()
```

## Plot WAXS together

```python
fig,axes=create_fig_ax( H=1,V=1, figsize=[8,6], title='',fontsize=12, y=1.08  ) 

i=0
#for sam in [  sams[0], sams[1]  ]:
for sam in sams:    
    ca = qiqw[sam]
    q, iq = ca[:,0], ca[:,1] + i #*1e3

    ax = axes[0]   
    plot1D( x= q, y= iq  , m= markers[i], ls='-', ax=ax, c=colors[i], markersize=1,
        logy=True, logx= False, 
        title='circular average-WAXS', 
       legend='sam=%s'%sam[7:], fontsize=16, tick_size =18, 
       xlabel=r'$Q (\AA^{-1})$', ylabel= r'$Intensity (a.u.)$', )
    
    xm,ym =  find_max_peak( q, iq , xrange=[0.2, 0.4 ])
    La = get_La( xm )
    qs = get_2dhx_qs( m = 8, a = La )
    print(qs[0], qs[-1], La)
    
    #ax.set_xlim([0.12, 4 ])   
    #ax.set_ylim([  1e4, ym*5.0])
    if i ==100:
        ax.vlines( qs, 0, ym*5 , label=r'$2D\ hexagonal\ L_a=%.2f\ nm$'%(La), 
                  linestyle='-' , linewidth=.5  )
  
    
    
    i += 1
ax.legend(loc = 'best', fontsize= 12 )
ax.set_ylabel( r'$Intensity (a.u.)$', fontsize=18)
ax.set_xlabel( r'$Q (\AA^{-1})$', fontsize=18)
    
```
