# A loop to sort

```python
ssam_dict = {}    
for sam in ks:     
    print( 'The sample is: %s.'%sam)
    flist = np.sort( sam_dict[ sam ] )
    sflistd, sflist  = sort_fnlist( flist, pattern='', cut_beg = -15, cut_end=-10, method='cut_from_end' )
    pairs = get_ytiling_pairs( sflist, sid_pos = [-15, -10 ],  pos12_gap = 7 ) 
    print('There are %s pairs to be stitched.'%( len( pairs.keys() ) ))
    print('The filename list for this sam=%s are:\n %s.'%(sam, pairs ))
    ssam_dict[sam] = pairs

    
    
    
```

# For stitching

```python
if scat == 'SAXS':
    SC = CMS_Scattering(  {  "det_str":'Pilatus2m' }  )   
if scat == 'WAXS':
    SC = CMS_Scattering(  {  "det_str":'Pilatus800' }  )  

 
```

```python
ks[-2:-1]
```

```python
for sam in ks[:]:  
#for sam in ks[-2:-1]:     
    pairk = ssam_dict[sam]
    ksk = list( pairk.keys())
    for pi in ksk[:] :
        fp12 = pairk[pi]
        print( fp12 )    
        full_filenames = [ inDir + fp for fp in  fp12 ] 
        assert len(full_filenames) == 2        
        SC.load_data( full_filenames ) 
        img_st = SC.stitich( img1 = SC.imgs[0], img2 =SC.imgs[1], det=SC.det_str)          
        filename = fp12[0][:-10] + '_stitched' + fp12[0][-10:]
        save_array_to_tiff( img_st, outDir + filename, verbose=True )
```
